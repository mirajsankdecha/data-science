```python
import re
```

```python
text = """hello world!!! python is a beautiful language!
I have started learning python in 2005.
Here is my lucky number 8.
here is my email:email@email.com and info@gptalkie.com and 
udemy@kgptalkie.com"""
```

```python
pattern = 'hello'
result = re.match(pattern=pattern, string=text)
span = result.span()
text[span[0] : span[1]]
```

```python
pattern = r'python'
result = re.match(pattern=pattern, string=text)
print(result)
```

```python
pattern = r'python'
result = re.search(pattern=pattern, string=text)
print(result)
```

```python
pattern = r'\d+'
result = re.search(pattern=pattern, string=text)
print(result)
```

```python
pattern = r'\w+@\w+.\w+'
result = re.search(pattern=pattern, string=text)
print(result)
```

```python
pattern = r'\w+@\w+.\w+'
result = re.findall(pattern=pattern, string=text)
print(result)
```

```python
pattern = r'my'
result = re.findall(pattern=pattern, string=text)
print(result)
```

```python
pattern = r'\d{4}-\d{2}-\d{2}'
result = re.findall(pattern=pattern, string="today's date is 2026-09-01")
print(result)
```

```python
text = "8511411930"
pattern = r'0?\d{10}'
re.findall(pattern, text)
```

```python
text = "8511411930"
pattern = r'0?\d{10}'
result = re.sub(pattern, "*", text)
print(result)
```

```python
text = """hello world!!! python is a beautiful language!
I have started learning python in 2005.
Here is my lucky number 8.
here is my email:email@email.com and info@gptalkie.com and 
udemy@kgptalkie.com"""

def replace_with_star(match):
    return '*' * len(match.group())

pattern = r'0?d{10}'
result = re.sub(pattern, replace_with_star, text)
print(result)
```
