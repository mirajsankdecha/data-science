```python
from langchain_ollama import ChatOllama

base_url = "http://localhost:11434"
model = 'llama3.2'

llm = ChatOllama(
    base_url=base_url,
    model = model,
    temperature = 0.3,
    num_predict = 256
)
response = llm.invoke('write about india in 3 sentence')

print(response.content)
```
