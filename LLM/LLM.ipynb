```python
from langchain_ollama import ChatOllama

base_url = "http://localhost:11434"
model = 'llama3.2'

llm = ChatOllama(
    base_url=base_url,
    model = model,
    temperature = 0.3,
    num_predict = 1000
)
response = llm.invoke('write about India')

print(response.content)

# print(response.response_metadata) # See the mesuarments of HTTP Response
```

```python
ans = response.content
words = ans.split(" ")
approx_tokens = len(words)
print(approx_tokens) # This is rounghly
```

```python
response = ""
for chunk in llm.stream('write about india in 3 sentence'):
    response = response + " " + chunk.content
    print(response)
```

```python
from langchain_core.messages import SystemMessage, HumanMessage

base_url = "http://localhost:11434"
model = 'llama3.2'

llm = ChatOllama(base_url=base_url, model=model)
```

```python
question = HumanMessage('write about india in 3 sentence')
system = SystemMessage('You are a diplomat . You answer in short sentences.')

messages = [system, question]
response = llm.invoke(messages)

print(response.content)
```

```python
question = HumanMessage('write about india in 3 sentence')
system = SystemMessage('You are phd teacher. You answer in short sentences.')

messages = [system, question]
response = llm.invoke(messages)

print(response.content)
```

```python
from langchain_core.prompts import (
                                        SystemMessagePromptTemplate,
                                        HumanMessagePromptTemplate,
                                        PromptTemplate,
                                        ChatPromptTemplate
                                        )
```

```python
system = SystemMessagePromptTemplate.from_template('You are {school} teacher. You answer in short sentences.')

question = HumanMessagePromptTemplate.from_template('tell me about the {topics} in {points} points')
```

```python
messages = [system,question]
template = ChatPromptTemplate(messages)
question = template.invoke({'school': 'elementary', 'topics': 'sun', 'points': 5})

response = llm.invoke(question)
print(response.content)
```
